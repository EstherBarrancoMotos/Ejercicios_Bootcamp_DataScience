## Building a regression model

The data have already been split to predictor (*data*) and response (*target*) variables. Given this information, we'll apply what we have previously learned about linear regression and apply the algorithm to the diabetes dataset.

#### In the cell below, import the `linear_model` class from `sklearn`. 

```python
from sklearn.linear_model import LinearRegression
```

#### Create a new instance of the linear regression model and assign the new instance to the variable `diabetes_model`.

```python
diabetes_model = LinearRegression()
```

#### Next, let's split the training and test data.

Define `diabetes_data_train`, `diabetes_target_train`, `diabetes_data_test`, and `diabetes_target_test`. Use the last 20 records for the test data and the rest for the training data.

```python
diabetes_data_train = diabetes.data[:-20]
diabetes_data_test = diabetes.data[-20:]

diabetes_target_train = diabetes.target[:-20]
diabetes_target_test = diabetes.target[-20:]
```

```python
print(diabetes_data_train.shape)
print(diabetes_data_test.shape)

print(diabetes_target_train.shape)
print(diabetes_target_test.shape)
```

```python
print(diabetes_data_train)
print(diabetes_data_test)

print(diabetes_target_train)
print(diabetes_target_test)
```

Perform a brief exploratory analysis to investigate the conditions necessary for linear regression. This analysis will help you understand the distribution of the target variable and its correlations with other features.

Fit the training data and target to `diabetes_model`. Print the *intercept* and *coefficients* of the model.

```python
from sklearn.linear_model import LinearRegression

diabetes_model = LinearRegression()

diabetes_model.fit(diabetes_data_train, diabetes_target_train)

print("Intercepto:", diabetes_model.intercept_)
print("Coeficientes:", diabetes_model.coef_)
```

```python
diabetes_model.fit(diabetes_data_train, diabetes_target_train, sample_weight=42)
diabetes_model
```

#### Inspecting the results

From the outputs you should have seen:

- The intercept is a float number.
- The coefficients are an array containing 10 float numbers.

This is the linear regression model fitted to your training dataset.

#### Is `diabetes_target_test` exactly the same as the model prediction? Use different error metrics

```python
from sklearn.metrics import r2_score

prediction_train = diabetes_model.predict(diabetes_data_train)
prediction_test = diabetes_model.predict(diabetes_data_test)

r2_score_train = r2_score(diabetes_target_train, prediction_train)
r2_score_test = r2_score(diabetes_target_test, prediction_test)

print("R2 train:", r2_score_train)
print("R2 test:", r2_score_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

prediction = diabetes_model.predict(diabetes_data_test)

mae = mean_absolute_error(diabetes_target_test, prediction)
mse = mean_squared_error(diabetes_target_test, prediction)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((diabetes_target_test - prediction) / diabetes_target_test))

print("mean_absolute_error", mae)
print("mean_squared_error", mse)
print("root_mean_squared_error", rmse)
print("mean_absolute_percentage_error", mape)
```

```python
import matplotlib.pyplot as plt

# crear errores
errores = diabetes_target_test - prediction

# histograma
plt.hist(errores, bins=10, edgecolor='black')
plt.xlabel("target")
plt.ylabel("Count")
plt.title("Histogram of errors")

# línea en 0 (opcional pero pro)
plt.axvline(0)

plt.show()
```
